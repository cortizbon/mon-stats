```python
import pandas as pd
```

```python
inf_empl = pd.read_csv('inf_empl.csv')
tir_tib = pd.read_csv('tir_tib.csv')
```

```python
inf_empl
```

```python
tir_inf = pd.merge(inf_empl[['fecha','Inflación total', 'Tasa de desempleo (%)', 'Meta de inflación']], 
         tir_tib[['Fecha', 'TIR']], left_on='fecha', right_on='Fecha').iloc[::-1].reset_index(drop=True).drop(columns='Fecha')
```

```python
tir_inf.to_csv('tir_inf.csv', index=False)
```
